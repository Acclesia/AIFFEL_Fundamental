# 컴퓨터는 크게 하드웨어와 소프트웨어로 이뤄져 있다. 

### 하드웨어 
하드웨어에는 메인보드가 있는데 그 안에는 CPU, 주기억장치(RAM), 보조기억장치(하드디스트)가 있다. 
실제 데이터를 처리하는 공간임으로 백엔드라고도 한다.

### 역할
CPU는 실행
주기억장치(메모리): 휘발성 정보 저장
보조기억장치(하드디스크): 비휘발성 정보 저장

하드디스크 안에는 운영체제가 있다
운영체제는 제어하는 역할을 한다.
컴퓨터에서는 CPU, MEM, DIC 정보가 오가는데 이를 제어하고 어떤 것부터 일을 처리할지 조절하는 역할이다.
예를 들어 리눅스, 맥, 윈도우 등이 있다.

* 사용자가 컴퓨터를 키면 운영체제에서 보조기억장치에 저장된 데이터를 처리할지 결정한 후 메모리에 UPLOAD되고 CPU가 실행된다.

### 소프트웨어
소프트웨어는 우리가 만질 수 없는 것이며 하드웨어가 작동하도록 명령을 내린다. 

화면에 보이는 작업이기에 프론트엔드라고도 한다.

소프트웨어는 두가지 종류가 있다.
1. 시스템 소프트웨어: 윈도우, 리눅스, 우분투 같은 것들로 컴퓨터 작동을 위해 존재한다.
2. 응용 소프트웨어: 파워포인트,앱,엑셀,인터넷같은 것들이며 편리한 사용을 위해 존재한다.
