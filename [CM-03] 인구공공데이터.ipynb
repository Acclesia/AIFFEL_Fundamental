# 2. 다양한 형태로 시각화하기

```python
# 막대그래프 그리기
import csv

f = open('new_age.csv', 'r', encoding='cp949')
data = csv.reader(f)

result=[]
for row in data :
    if '신도림' in row[0]: #신도림이 있는 행정구역
        for i in row[3:]: # 0-100까지 모든 연령 반복
            result.append(int(i.replace(",",""))) #쉼표랑 인트로 바꿈

import matplotlib.pyplot as plt
plt.bar(range(101), result)
plt.show()
```

```python
plt.barh(range(101), result)
plt.show()
```

```python
# 항아리 모양 그래프 만들기 

f = open('gender.csv', 'r', encoding='cp949')
data = csv.reader(f)

m=[] #남자는 처으부터 차례대로 저장
f=[] #여자는 뒤에서 저장한 후 다시 역순으로 뒤집는다.

for row in data :
    if '신도림' in row[0]: #신도림이 있는 행정구역
        for i in row[3:104]:
            m.append(-int(i.replace(",","")))# 0-100까지 모든 연령 반복
        for i in row[106:]:
            f.append(int(i.replace(",","")))
```

```python
plt.barh(range(101),m)
plt.barh(range(101),f)
plt.show()
```

```python
# 항아리 모양 그래프 만들기 

f = open('gender.csv', 'r', encoding='cp949')
data = csv.reader(f)

m=[] #남자는 처으부터 차례대로 저장
f=[] #여자는 뒤에서 저장한 후 다시 역순으로 뒤집는다.

name= input("찾고 싶은 지역의 이름을 알려주세요 : ")
for row in data :
    if '신도림' in row[0]: #신도림이 있는 행정구역
        for i in row[3:104]:
            m.append(-int(i.replace(",","")))# 0-100까지 모든 연령 반복
        for i in row[106:]:
            f.append(int(i.replace(",","")))
        break    
plt.style.use('ggplot')
plt.figure(figsize=(10,5),dpi=300)
plt.rc('font',family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False
plt.title(name+'지역의 남녀 성별 인구 분포')            
plt.barh(range(101),m, label='male')
plt.barh(range(101),f, label='female')
plt.legend()
plt.show()

print(len(m), len(f))
```

# 3. 파이차트로 나타내기

```python
#pie chart

plt.pie([10,20])
plt.show()
```

```python
#pie chart 동그란 원

size=[2441,2312,1031,1233]
label=['A','B','C','D']

#plt.rc('font',family = 'Malgun Gothic')
plt.axis('equal') #동그란 원 그리기 
plt.pie(size, labels=label, autopct='%.1f%%') #항목별 비율을 표시하는 함수 autopct 
plt.legend()                # 소수점 첫번째까지 표현하고자 0.1까지라고 표시!
plt.show()
```

```python
#pie chart 파이의 색 변경하기, 돌출효과
# 돌출효과는 explode! 돌출되지 않음은 0 이다

size=[2441,2312,1031,1233]
label=['A','B','C','D']
color = ['red','deeppink','hotpink','pink']
#plt.rc('font',family = 'Malgun Gothic')
plt.axis('equal') #동그란 원 그리기 
plt.pie(size, labels=label, autopct='%.1f%%',colors=color,explode=(0,0,0.2,0)) #항목별 비율을 표시하는 함수 autopct 
plt.legend()                # 소수점 첫번째까지 표현하고자 0.1까지라고 표시!
plt.show()
```

```python
#제주도의 성별인구 비율 표현하기 
f = open('gender.csv', 'r', encoding='cp949')
data = csv.reader(f)

size=[] # 남여 성별 인구의 합계

name= input("찾고 싶은 지역의 이름을 알려주세요 : ")
for row in data :
    if name in row[0]: #신도림이 있는 행정구역
        m=0
        f=0
        for i in range(101): #0-100세까지의 인구
            m += int(row[i+3].replace(",","")) #남자 연령별 인구 3-103까지 데이터를 m에 더하기 
            f += int(row[i+106].replace(",","")) #여자 연령벼인구 106-206
        break
size.append(m) #연령별 인구를 모두 더해서 합게 데이터 리스트에 추가 
size.append(f)
print(size) #남녀 총 인구별 데이터
```

# 꺽은선그래프로 나타내기 

```python
f = open('gender.csv', 'r', encoding='cp949')
data = csv.reader(f)
m=[]
f=[]
name=input('궁금한 동네를 입력해주세요: ')

for row in data:
    if name in row[0]:
        for i in range(3,104):
            m.append(int(row[i].replace(",","")))
            f.append(int(row[i+103].replace(",","")))
```

```python
plt.plot(m, label='Male')
plt.plot(f, label='Female')
plt.legend()
plt.show()
```

```python
#막대그래프로 표현하기 
f = open('gender.csv', 'r', encoding='cp949')
data = csv.reader(f)
result=[]
name=input('궁금한 동네르 입력해주세요: ')

for row in data:
    if name in row[0]:
        for i in range(3,104):
            result.append(int(row[i].replace(",","")-int(row[i+103].replace(",",""))))
        break
```

```python
plt.bar(range(101),result)
plt.show()
```

```python
import random
x=[]
y=[]
size=[]

for i in range(100): #100개의 데이터를 50-100범위에서 scatter
    x.append(random.randint(50,100))
    y.append(random.randint(50,100))
    size.append(random.randint(50,100))
    
plt.scatter(x,y,s=size,c=size,cmap='jet',alpha=0.7) #알파속성으로 투명도 조절
plt.colorbar()
plt.show()
    
```

```python
# 제주도 연령대별 성별비율을 산점도로 표현하기
f = open('gender.csv', 'r', encoding='cp949')
data = csv.reader(f)

m=[]
f=[]
name= input("찾고 싶은 지역의 이름을 알려주세요 : ")
for row in data :
    if name in row[0]: #신도림이 있는 행정구역

        for i in range(3,104):
            m.append(int(row[i].replace(",","")))
            f.append(int(row[i+103].replace(",","")))
        break

        
plt.style.use('ggplot')
plt.figure(figsize=(10,5),dpi=300)
plt.title(name+'local gender grap')
plt.scatter(m,f, c=range(101),alpha=0.7,cmap='jet') # 0-100세까지 여성과 남성별 인구수 
plt.colorbar()
plt.plot(range(max(m)),range(max(m)),'g') #중간 추세선
plt.xlabel('male population')
plt.ylabel('female population')
plt.show()
```
